# Median

```python

def calculate_median(arr):
    """Calculate the median of a numeric array."""
    if len(arr) == 0:
        raise ValueError("Input array must not be empty.")
    sorted_arr = np.sort(arr)
    n = len(sorted_arr)
    mid = n // 2
    if n % 2 == 1:
        return float(sorted_arr[mid])
    else:
        return float((sorted_arr[mid - 1] + sorted_arr[mid]) / 2)
```

# Mode

```python

def calculate_mode(arr):
    """
    Calculate the mode(s) of a dataset.
    Returns a list of mode(s). Raises warning if no unique mode.
    """
    if len(arr) == 0:
        raise ValueError("Input array must not be empty.")
    
    # Use scipy.stats.mode for efficiency and handling edge cases
    mode_result = stats.mode(arr, keepdims=True)
    mode_values = mode_result.mode
    counts = mode_result.count

    if counts[0] == 1 and len(np.unique(arr)) == len(arr):
        # All values are unique
        return []  # No mode
    else:
        # Return all values that share the max frequency
        unique_vals, counts_ = np.unique(arr, return_counts=True)
        max_count = np.max(counts_)
        modes = unique_vals[counts_ == max_count].tolist()
        return modes
```
