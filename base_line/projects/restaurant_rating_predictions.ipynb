## Validation testing

### MSE && RMSE

```python
def mse(y_test, y_pred):
    """
    Mean Squared Error (MSE)

    MSE mengukur rata-rata kuadrat selisih antara nilai aktual (y_test)
    dan prediksi model (y_pred).

    Secara matematis:
        MSE = (1/n) * Σ (y_i - ŷ_i)^2

    Interpretasi Model:
    -------------------
    1. Mengukur "seberapa jauh" prediksi model dari nilai sebenarnya.
    2. Error dikuadratkan → kesalahan besar dihukum lebih berat.
    3. Sangat sensitif terhadap outlier karena efek kuadrat.
    4. Dalam regresi linear, meminimalkan MSE sama dengan mencari
       estimator Maximum Likelihood jika error ~ Normal(0, σ²).

    Makna Statistik:
    ----------------
    - MSE = estimasi varians error model.
    - Jika MSE kecil → model memiliki prediksi yang presisi.
    - Jika MSE besar → model memiliki deviasi tinggi terhadap data nyata.

    Sifat Penting:
    --------------
    - Differentiable → cocok untuk optimisasi gradient descent.
    - Convex untuk model linear → solusi global unik.
    """
    test = np.array(y_test)
    pred = np.array(y_pred)
    return np.mean((test - pred) ** 2)


def rmse(y_test, y_pred):
    """
    Root Mean Squared Error (RMSE)

    RMSE adalah akar dari MSE.

    Secara matematis:
        RMSE = sqrt(MSE)

    Interpretasi Model:
    -------------------
    1. Mengukur rata-rata besar error dalam satuan asli target.
       (Jika target dalam juta rupiah → RMSE juga juta rupiah.)
    2. Lebih mudah dipahami secara praktis dibanding MSE.
    3. Tetap sensitif terhadap outlier karena berbasis MSE.

    Makna Statistik:
    ----------------
    - RMSE ≈ standar deviasi residual (error model).
    - Memberikan gambaran "seberapa besar kesalahan tipikal model".

    Perbandingan dengan MSE:
    ------------------------
    - MSE → unit kuadrat (kurang intuitif).
    - RMSE → unit asli (lebih interpretatif).
    - Dalam optimisasi, biasanya tetap minimisasi MSE,
      lalu laporkan RMSE untuk interpretasi.

    Jika RMSE = 5:
    Artinya rata-rata model meleset sekitar ±5 unit dari nilai aktual.
    """
    return mse(y_test, y_pred) ** 0.5
```

### MAE

```python
def mae(y_test, y_pred):
    """
    Mean Absolute Error (MAE)

    MAE mengukur rata-rata nilai absolut dari selisih antara
    nilai aktual (y_test) dan prediksi model (y_pred).

    Secara matematis:
        MAE = (1/n) * Σ |y_i - ŷ_i|

    Interpretasi Model:
    -------------------
    1. Mengukur besar rata-rata kesalahan model tanpa memperhatikan arah.
    2. Semua error diperlakukan secara linear (tidak dikuadratkan).
    3. Tidak terlalu menghukum error besar dibanding MSE.
    4. Lebih robust terhadap outlier dibanding MSE/RMSE.

    Makna Statistik:
    ----------------
    - MAE meminimalkan median residual.
    - Jika kita meminimalkan MAE, solusi optimal adalah median (bukan mean).
    - Cocok ketika distribusi error tidak normal atau heavy-tailed.

    Hubungan Probabilistik:
    -----------------------
    Jika diasumsikan error mengikuti distribusi Laplace:
        ε ~ Laplace(0, b)

    Maka meminimalkan MAE ≡ memaksimalkan likelihood distribusi Laplace.

    Perbandingan dengan MSE:
    ------------------------
    - MSE menghukum error besar secara kuadrat.
    - MAE menghukum error besar secara linear.
    - MSE lebih sensitif terhadap outlier.
    - MAE lebih stabil jika data mengandung noise ekstrem.

    Interpretasi Praktis:
    ---------------------
    Jika MAE = 5:
        Artinya model rata-rata meleset ±5 unit dari nilai aktual.

    Sifat Matematis:
    ----------------
    - Tidak differentiable di titik 0 (karena absolute value).
    - Optimisasi biasanya menggunakan subgradient.
    """
    test=np.array(y_test)
    pred=np.array(y_pred)

    return np.mean(np.abs(test - pred))
```

### R^2

```python
def R2(y_test, y_pred):
    """
    R-squared (Coefficient of Determination)

    Mengukur proporsi variasi dalam data aktual (y_test)
    yang dapat dijelaskan oleh model (y_pred).

    Rumus:
        R² = 1 - (SS_res / SS_tot)

    dimana:
        SS_res = Σ (y_i - ŷ_i)^2     → Residual Sum of Squares
        SS_tot = Σ (y_i - ȳ)^2       → Total Sum of Squares
        ȳ      = rata-rata y_test

    Interpretasi Statistik:
    -----------------------
    - SS_tot  = total variasi dalam data.
    - SS_res  = variasi yang tidak bisa dijelaskan model.
    - R²      = proporsi variasi yang berhasil dijelaskan model.

    Secara ekuivalen:
        R² = (Varians yang dijelaskan) / (Total varians)

    Interpretasi Nilai:
    -------------------
    R² = 1   → Model sempurna (tidak ada error)
    R² = 0   → Model sama dengan menebak rata-rata
    R² < 0   → Model lebih buruk dari rata-rata

    Hubungan dengan MSE:
    --------------------
    Karena:
        SS_res = n * MSE

    Maka R² bisa dilihat sebagai:
        R² = 1 - (MSE / Var(y))

    Artinya:
    - Jika MSE kecil → R² mendekati 1.
    - Jika MSE besar → R² mendekati 0 atau negatif.

    Makna Konseptual:
    -----------------
    R² menjawab pertanyaan:
        "Seberapa besar variasi data yang bisa dijelaskan model?"

    Catatan Penting:
    ----------------
    - R² tidak menunjukkan apakah model bias.
    - Tidak menjamin model baik secara prediktif.
    - Selalu meningkat jika fitur ditambah (pada regresi biasa).
    - Untuk penalti kompleksitas gunakan Adjusted R².
    """

    test = np.array(y_test)
    pred = np.array(y_pred)

    mean_y_test = np.mean(test)

    ss_residual = np.sum((test - pred) ** 2)
    ss_total = np.sum((test - mean_y_test) ** 2)

    return 1 - (ss_residual / ss_total)
```

### adjusted R2

```python
def adjusted_R2(y_test, y_pred, p_fitur):
    """
    Adjusted R-squared (R² yang disesuaikan)

    Mengukur proporsi variasi dalam data yang dapat dijelaskan model,
    dengan penalti terhadap jumlah fitur (kompleksitas model).

    Parameter:
    ----------
    y_test : array-like
        Nilai aktual.
    y_pred : array-like
        Nilai prediksi model.
    p_fitur : int
        Jumlah fitur (predictor) dalam model.

    Rumus:
    ------
        R²_adj = 1 - ((1 - R²) * (n - 1)) / (n - p - 1)

    dimana:
        n  = jumlah observasi
        p  = jumlah fitur
        R² = koefisien determinasi biasa

    Interpretasi Statistik:
    -----------------------
    - R² biasa selalu meningkat ketika fitur ditambahkan.
    - Adjusted R² mengoreksi hal tersebut dengan memperhitungkan
      derajat kebebasan (degrees of freedom).

    Secara ekuivalen:
        R²_adj =
        1 - (SS_res / (n - p - 1)) /
            (SS_tot / (n - 1))

    Artinya:
        R²_adj membandingkan varians error yang sudah dikoreksi
        terhadap varians total data.

    Interpretasi Nilai:
    -------------------
    R²_adj mendekati 1  → model menjelaskan variasi dengan baik
    R²_adj ≈ 0         → model hampir tidak lebih baik dari mean
    R²_adj < 0         → model lebih buruk dari mean

    Perbedaan dengan R²:
    --------------------
    - R² selalu naik jika fitur ditambah.
    - R²_adj bisa naik atau turun.
    - Jika fitur baru tidak signifikan → R²_adj turun.

    Makna Praktis:
    --------------
    Digunakan untuk:
        - Membandingkan model dengan jumlah fitur berbeda
        - Mendeteksi overfitting
        - Model selection dalam regresi linear

    Catatan:
    --------
    - Berlaku terutama untuk regresi linear OLS.
    - Tidak cocok langsung untuk model non-linear kompleks.
    """

    r2 = R2(y_test, y_pred)
    n = len(y_test)

    return 1 - (((1 - r2) * (n - 1)) / (n - p_fitur - 1))
```

### Metric evaluation

```python
def metric_eval(y_test, y_pred):
    mse_val = mse(y_test, y_pred)
    rmse_val = rmse(y_test, y_pred)
    mae_val = mae(y_test, y_pred)
    r2_val = R2(y_test, y_pred)

    print(f"MSE : {mse_val:.4f}")
    print(f"RMSE: {rmse_val:.4f}")
    print(f"MAE : {mae_val:.4f}")
    print(f"R²  : {r2_val:.4f}")
```

## Bias Variance tradeoff

```python
def bias(y_test, y_pred):
    """
    Bias (Mean Error)

    Mengukur rata-rata selisih antara nilai aktual dan prediksi.

    Rumus:
        Bias = (1/n) * Σ (y_i - ŷ_i)

    Interpretasi:
    -------------
    - Bias > 0  → model cenderung underpredict
                  (prediksi lebih kecil dari nilai aktual)
    - Bias < 0  → model cenderung overpredict
                  (prediksi lebih besar dari nilai aktual)
    - Bias ≈ 0  → tidak ada kecenderungan sistematis

    Catatan:
    --------
    Bias kecil tidak berarti model bagus.
    Bisa saja error besar tapi saling menghapus.
    """

    test = np.array(y_test)
    pred = np.array(y_pred)

    return np.mean(test - pred)
```

```python
def variance(y_test, y_pred):
    """
    Menghitung variance (ragam) dari residual / error prediksi model.

    Residual didefinisikan sebagai:
        e_i = y_i - ŷ_i

    Fungsi ini menghitung population variance dari error:

        Var(e) = (1/n) * Σ (e_i - ē)^2

    di mana:
        e_i  = residual ke-i
        ē    = rata-rata residual
        n    = jumlah sampel

    Interpretasi:
    -------------
    - Mengukur seberapa menyebar error model terhadap rata-ratanya.
    - Nilai kecil  → error konsisten (model stabil).
    - Nilai besar → error fluktuatif (model kurang stabil).

    Hubungan dengan MSE:
    --------------------
        MSE = Variance(error) + (Mean Error)^2

    Jika mean error = 0 (tidak bias), maka:
        MSE = Variance(error)

    Catatan penting:
    ----------------
    - Ini adalah variance residual pada satu dataset.
    - Ini BUKAN variance model dalam bias-variance tradeoff teoritis.
    - Variance model membutuhkan training berulang pada dataset berbeda.

    Parameters:
    -----------
    y_test : array-like
        Nilai aktual (ground truth).

    y_pred : array-like
        Nilai prediksi dari model.

    Returns:
    --------
    float
        Nilai variance dari residual.
    """
    test=np.array(y_test)
    pred=np.array(y_pred)

    error=test-pred
    return np.sum(( error - np.mean(error))**2)/len(error)
```

## Training technique

```python
def train_test_split(x, y, test_size=0.2, shuffle=True, random_state=None):
    """
    Membagi dataset menjadi data training dan data testing.

    Parameters
    ----------
    x : array-like
        Fitur / variabel independen.
    y : array-like
        Target / variabel dependen.
    test_size : float, default=0.2
        Proporsi data yang digunakan sebagai data testing.
    shuffle : bool, default=True
        Jika True, data akan diacak sebelum dibagi.
    random_state : int atau None
        Seed untuk random agar hasil pembagian reproducible.

    Returns
    -------
    x_train, x_test, y_train, y_test : tuple of arrays
        Dataset yang sudah dibagi.
    """
    
    x = np.array(x)
    y = np.array(y)

    n = len(x)
    test_count = int(n * test_size)

    if shuffle:
        if random_state is not None:
            np.random.seed(random_state)
        indices = np.random.permutation(n)
        x = x[indices]
        y = y[indices]

    x_test = x[:test_count]
    y_test = y[:test_count]

    x_train = x[test_count:]
    y_train = y[test_count:]

    return x_train, x_test, y_train, y_test
```

```python

def k_fold(x, y, k=5, shuffle=True, random_state=None):
    """
    Melakukan K-Fold Cross Validation splitting.

    Dataset dibagi menjadi K bagian (fold). Pada setiap iterasi:
    - 1 fold digunakan sebagai data testing
    - K-1 fold digunakan sebagai data training.

    Parameters
    ----------
    x : array-like
        Fitur / variabel independen.
    y : array-like
        Target / variabel dependen.
    k : int
        Jumlah fold.
    shuffle : bool
        Apakah data diacak sebelum dibagi.
    random_state : int atau None
        Seed random agar reproducible.

    Returns
    -------
    folds : list of tuples
        List berisi (x_train, x_test, y_train, y_test).
    """

    x = np.array(x)
    y = np.array(y)

    n = len(x)

    if shuffle:
        if random_state is not None:
            np.random.seed(random_state)
        indices = np.random.permutation(n)
        x = x[indices]
        y = y[indices]

    fold_size = n // k
    folds = []

    for i in range(k):
        start = i * fold_size
        end = start + fold_size if i != k-1 else n

        x_test = x[start:end]
        y_test = y[start:end]

        x_train = np.concatenate((x[:start], x[end:]))
        y_train = np.concatenate((y[:start], y[end:]))

        folds.append((x_train, x_test, y_train, y_test))

    return folds
```

```python
def stratified_k_fold(df, target_col, k=5, shuffle=True, random_state=None):
    """
    Melakukan Stratified K-Fold Cross Validation.

    Setiap fold mempertahankan proporsi kelas target yang sama
    seperti pada dataset asli.

    Parameters
    ----------
    df : pandas.DataFrame
        Dataset lengkap.
    target_col : str
        Nama kolom target.
    k : int
        Jumlah fold.
    shuffle : bool
        Apakah data diacak sebelum pembagian.
    random_state : int atau None
        Seed untuk reproducibility.

    Returns
    -------
    folds : list of tuples
        List berisi (train_df, test_df) untuk setiap fold.

    Interpretasi
    ------------
    - Cocok untuk klasifikasi dengan data imbalance.
    - Mengurangi variance evaluasi performa.
    - Setiap fold memiliki distribusi kelas yang seimbang.
    """

    if shuffle:
        df = df.sample(frac=1, random_state=random_state).reset_index(drop=True)

    folds = [[] for _ in range(k)]

    # Kelompokkan berdasarkan kelas
    grouped = df.groupby(target_col)

    for _, group in grouped:
        group = group.reset_index(drop=True)
        n = len(group)
        fold_size = n // k

        for i in range(k):
            start = i * fold_size
            end = start + fold_size if i != k-1 else n
            folds[i].append(group.iloc[start:end])

    stratified_folds = []

    for i in range(k):
        test_df = pd.concat(folds[i])
        train_df = pd.concat([pd.concat(folds[j]) for j in range(k) if j != i])

        stratified_folds.append((train_df, test_df))

    return stratified_folds
```

## standarization

```python
class StandardScaler:
    """
    Implementasi sederhana StandardScaler (mirip sklearn) dari nol.

    Melakukan Z-score normalization:

        z = (x - mean) / std

    Attributes
    ----------
    mean_ : numpy.ndarray
        Mean tiap fitur (kolom)

    scale_ : numpy.ndarray
        Standard deviation tiap fitur (kolom)
    """

    def __init__(self):
        self.mean_ = None
        self.scale_ = None

    def fit(self, X):
        """
        Menghitung mean dan standard deviation dari data.

        Parameters
        ----------
        X : array-like (n_samples, n_features)

        Returns
        -------
        self
        """
        X = np.array(X)

        if X.ndim == 1:
            X = X.reshape(-1, 1)

        self.mean_ = np.mean(X, axis=0)
        self.scale_ = np.std(X, axis=0)

        self.scale_[self.scale_ == 0] = 1

        return self

    def transform(self, X):
        """
        Melakukan standardization menggunakan mean dan std yang sudah di-fit.

        Parameters
        ----------
        X : array-like

        Returns
        -------
        numpy.ndarray
        """
        if self.mean_ is None or self.scale_ is None:
            raise ValueError("Model belum di-fit. Jalankan fit() terlebih dahulu.")

        X = np.array(X)

        if X.ndim == 1:
            X = X.reshape(-1, 1)

        return (X - self.mean_) / self.scale_

    def fit_transform(self, X):
        """
        Fit ke data lalu langsung transform.

        Equivalent dengan:
            scaler.fit(X)
            scaler.transform(X)
        """
        return self.fit(X).transform(X)
```

```python
class RobustScaler:
    """
    Implementasi sederhana RobustScaler (mirip sklearn).

    Transformasi:
        z = (x - median) / IQR

    di mana:
        IQR = Q3 - Q1

    Attributes
    ----------
    center_ : numpy.ndarray
        Median tiap fitur (kolom)

    scale_ : numpy.ndarray
        IQR tiap fitur (kolom)
    """

    def __init__(self):
        self.center_ = None
        self.scale_ = None

    def fit(self, X):
        """
        Menghitung median dan IQR dari data.

        Parameters
        ----------
        X : array-like (n_samples, n_features)

        Returns
        -------
        self
        """
        X = np.array(X)

        # Jika 1D, ubah jadi 2D
        if X.ndim == 1:
            X = X.reshape(-1, 1)

        self.center_ = np.median(X, axis=0)

        q1 = np.quantile(X, 0.25, axis=0)
        q3 = np.quantile(X, 0.75, axis=0)

        self.scale_ = q3 - q1

        # Hindari division by zero
        self.scale_[self.scale_ == 0] = 1

        return self

    def transform(self, X):
        """
        Melakukan robust scaling menggunakan median & IQR hasil fit.

        Parameters
        ----------
        X : array-like

        Returns
        -------
        numpy.ndarray
        """
        if self.center_ is None or self.scale_ is None:
            raise ValueError("Model belum di-fit. Jalankan fit() terlebih dahulu.")

        X = np.array(X)

        if X.ndim == 1:
            X = X.reshape(-1, 1)

        return (X - self.center_) / self.scale_

    def fit_transform(self, X):
        """
        Fit ke data lalu langsung transform.
        """
        return self.fit(X).transform(X)
```

## Simple Linear Regression

```python
class SimpleLinearRegression:
    """
    Simple Linear Regression (1 fitur) dari scratch.

    Model linear:
        y = w*x + b

    Rumus weight (w) dan bias (b):
        w = Σ((x - mean_x)*(y - mean_y)) / Σ((x - mean_x)^2)
        b = mean_y - w * mean_x

    Parameters
    ----------
    fit_intercept : bool, default=True
        Jika True, hitung bias (b). Jika False, b = 0.
    """

    def __init__(self, fit_intercept=True):
        self.weight = None
        self.bias = None
        self.fit_intercept = fit_intercept

    def fit(self, X, Y):
        x = np.array(X).flatten()
        y = np.array(Y).flatten()

        mean_x = np.mean(x)
        mean_y = np.mean(y)

        self.weight = np.sum((x - mean_x) * (y - mean_y)) / np.sum((x - mean_x)**2)

        if self.fit_intercept:
            self.bias = mean_y - self.weight * mean_x
        else:
            self.bias = 0.0

        return self

    def predict(self, X):
        x_test = np.array(X).flatten()
        return self.weight * x_test + self.bias
```

## Multiple Linear Regresstion

```python
class MultipleLinearRegression:
    """
    Multiple Linear Regression dari scratch (multi-fitur).

    Model:
        y = w1*x1 + w2*x2 + ... + wn*xn + b

    Closed form:
        w = (X^T X)^-1 X^T y
        b = intercept (ditambahkan ke y)
    """

    def __init__(self, fit_intercept=True):
        self.weight = None
        self.bias = None
        self.fit_intercept = fit_intercept

    def fit(self, x, y):
        x = np.array(x)
        y = np.array(y).flatten()

        if self.fit_intercept:
            ones = np.ones((x.shape[0], 1))
            x_b = np.hstack([ones, x])
        else:
            x_b = x

        w = np.linalg.pinv(x_b.T @ x_b) @ x_b.T @ y

        if self.fit_intercept:
            self.bias = w[0]
            self.weight = w[1:]
        else:
            self.bias = 0.0
            self.weight = w

        return self

    def predict(self, x_test):
        x_test = np.array(x_test)
        return x_test @ self.weight + self.bias
```

## Ridge Regression

```python
class RidgeRegression:
    """
    Ridge Regression (Linear Regression dengan regularisasi L2).

    Model ini mencari bobot w yang meminimalkan:
        ||y - Xw||^2 + alpha * ||w||^2

    Parameter
    ----------
    fit_intercept : bool, default=True
        Menentukan apakah model menghitung intercept (bias).

    alpha : float, default=1.0
        Kekuatan regularisasi L2.

    Atribut
    -------
    weight : ndarray of shape (n_features,)
        Koefisien model.

    bias : float
        Nilai intercept.
    """

    def __init__(self, fit_intercept: bool = True, alpha: float = 1.0) -> None:
        self.weight = None
        self.bias = None
        self.fit_intercept = fit_intercept
        self.alpha = alpha

    def fit(self, x, y):
        """
        Melatih model menggunakan data fitur x dan target y.

        Parameter
        ----------
        x : array-like of shape (n_samples, n_features)
            Data fitur.

        y : array-like of shape (n_samples,)
            Nilai target.
        """
        x = np.asarray(x)
        y = np.asarray(y).reshape(-1, 1)

        if self.fit_intercept:
            ones = np.ones((x.shape[0], 1))
            x_b = np.hstack((ones, x))
        else:
            x_b = x.copy()

        n_features = x_b.shape[1]
        I = np.eye(n_features)

        w = np.linalg.pinv(x_b.T @ x_b + self.alpha * I) @ x_b.T @ y

        if self.fit_intercept:
            self.bias = float(w[0])
            self.weight = w[1:].flatten()
        else:
            self.bias = 0.0
            self.weight = w.flatten()

    def predict(self, x_test):
        """
        Menghasilkan prediksi untuk data baru.

        Parameter
        ----------
        x_test : array-like of shape (n_samples, n_features)

        Returns
        -------
        ndarray of shape (n_samples,)
            Hasil prediksi.
        """
        x = np.asarray(x_test)
        return x @ self.weight + self.bias
```

## Lasso Regression

```python
class LassoRegression:
    """Lasso Regression dengan Coordinate Descent
    
    Mengimplementasikan Lasso (Least Absolute Shrinkage and Selection Operator)
    menggunakan algoritma Coordinate Descent untuk meminimalkan:
    
        (1/(2n)) * ||y - Xw - b||^2 + alpha * ||w||_1
    
    di mana:
        - ||w||_1 = sum(|w_j|) adalah L1 penalty
        - alpha mengontrol kekuatan regularisasi
        - Intercept (b) TIDAK diregularisasi
    
    Attributes:
        alpha (float): Kekuatan regularisasi L1. Nilai lebih besar → lebih banyak koefisien = 0.
            Default: 1.0
        fit_intercept (bool): Jika True, hitung intercept. Jika False, data diasumsikan 
            telah di-center. Default: True
        max_iter (int): Maksimum iterasi Coordinate Descent. Default: 1000
        tol (float): Toleransi konvergensi. Iterasi berhenti jika perubahan maksimum 
            koefisien < tol. Default: 1e-6
        weight (np.ndarray): Koefisien fitur setelah training. Bentuk (p,)
        bias (float): Intercept setelah training. 0.0 jika fit_intercept=False
    
    Notes:
        - **WAJIB standardisasi fitur** sebelum training untuk hasil optimal
        - Pada multikolinearitas sempurna, Lasso secara acak memilih salah satu fitur
        - Tidak memiliki solusi closed-form → HARUS menggunakan optimisasi iteratif
    
    Examples:
        >>> import numpy as np
        >>> from sklearn.preprocessing import StandardScaler
        >>> 
        >>> # Dataset dengan multikolinearitas (X2 = 2*X1)
        >>> X = np.array([[1, 2], [2, 4], [3, 6], [4, 8]])
        >>> y = np.array([3, 5, 7, 9])
        >>> 
        >>> # Standardisasi WAJIB untuk Lasso
        >>> scaler = StandardScaler()
        >>> X_scaled = scaler.fit_transform(X)
        >>> 
        >>> model = LassoRegression(alpha=1.0, max_iter=1000)
        >>> model.fit(X_scaled, y)
        Konvergen pada iterasi 5
        >>> 
        >>> print("Koefisien:", model.weight)
        Koefisien: [0.95 0.  ]  # X2 dieliminasi karena redundan
        >>> print("Intercept:", model.bias)
        Intercept: 6.0
        >>> print("Prediksi:", model.predict(X_scaled))
        Prediksi: [3.05 5.   6.95 8.9 ]
    """
    
    def __init__(self, alpha=1.0, fit_intercept=True, max_iter=1000, tol=1e-6):
        """Inisialisasi Lasso Regression
        
        Args:
            alpha (float): Kekuatan penalty L1. Harus >= 0.
                - alpha = 0 → setara dengan OLS (tanpa regularisasi)
                - alpha besar → solusi lebih sparse (banyak koefisien = 0)
            fit_intercept (bool): Jika True, hitung intercept secara terpisah 
                (tidak diregularisasi). Default: True
            max_iter (int): Maksimum iterasi Coordinate Descent. 
                Default: 1000
            tol (float): Toleransi untuk kriteria konvergensi. Iterasi berhenti 
                jika max(|w_new - w_old|) < tol. Default: 1e-6
        
        Raises:
            ValueError: Jika alpha < 0
        """
        if alpha < 0:
            raise ValueError(f"alpha harus >= 0, ditemukan {alpha}")
        
        self.alpha = alpha
        self.fit_intercept = fit_intercept
        self.max_iter = max_iter
        self.tol = tol
        self.weight = None
        self.bias = 0.0

    def __soft_threshold(self, rho, lam):
        """Operator soft-thresholding untuk Lasso
        
        Fungsi proximal dari penalty L1:
            S(rho, lam) = sign(rho) * max(|rho| - lam, 0)
        
        Args:
            rho (float): Partial correlation antara fitur dan residual
            lam (float): Threshold parameter (lambda/2)
        
        Returns:
            float: Nilai setelah soft-thresholding
        
        Examples:
            >>> model = LassoRegression()
            >>> model._LassoRegression__soft_threshold(3.0, 1.0)
            2.0
            >>> model._LassoRegression__soft_threshold(-3.0, 1.0)
            -2.0
            >>> model._LassoRegression__soft_threshold(0.5, 1.0)
            0.0
        """
        if rho > lam:
            return rho - lam
        elif rho < -lam:
            return rho + lam
        else:
            return 0.0

    def fit(self, X, y):
        """Fit model Lasso menggunakan Coordinate Descent
        
        Algoritma:
            1. Inisialisasi koefisien = 0
            2. Untuk setiap iterasi:
                a. Update intercept (jika fit_intercept=True)
                b. Untuk setiap fitur j:
                    - Hitung partial residual: r = y - Xw + w_j * X_j
                    - Hitung partial correlation: rho = X_j^T * r
                    - Update koefisien: w_j = S(rho, alpha/2) / ||X_j||^2
                c. Cek konvergensi: max|w_new - w_old| < tol
        
        Args:
            X (array-like): Matriks fitur dengan bentuk (n_samples, n_features)
            y (array-like): Vektor target dengan bentuk (n_samples,)
        
        Returns:
            self: Instance model yang telah di-fit
        
        Raises:
            ValueError: Jika X atau y memiliki bentuk tidak valid
            TypeError: Jika input tidak dapat dikonversi ke float
        
        Notes:
            - Input 1D untuk X akan otomatis diubah menjadi (n_samples, 1)
            - Intercept tidak pernah diregularisasi
            - Konvergensi dicetak ke stdout jika tercapai sebelum max_iter
        """
        # Validasi dan konversi input
        X = np.asarray(X, dtype=float)
        y = np.asarray(y, dtype=float).flatten()
        
        # Handle input 1D (vektor)
        if X.ndim == 1:
            X = X.reshape(-1, 1)
        
        if X.shape[0] != y.shape[0]:
            raise ValueError(
                f"Dimensi tidak sesuai: X punya {X.shape[0]} sampel, "
                f"y punya {y.shape[0]} sampel"
            )
        
        n, p = X.shape
        w = np.zeros(p)
        
        # Inisialisasi intercept
        if self.fit_intercept:
            b = np.mean(y)
        else:
            b = 0.0
        
        # Coordinate Descent
        for iteration in range(self.max_iter):
            w_old = w.copy()
            
            # Update intercept (tidak diregularisasi)
            if self.fit_intercept:
                b = np.mean(y - X @ w)
            
            # Update setiap koefisien fitur
            for j in range(p):
                # Hitung partial residual: residual tanpa kontribusi fitur j
                # r = y - (Xw - w_j * X_j) = y - Xw + w_j * X_j
                residual = y - (b + X @ w) + w[j] * X[:, j]
                
                # Hitung partial correlation dan norm kuadrat fitur
                rho = np.dot(X[:, j], residual)
                z = np.dot(X[:, j], X[:, j])  # = ||X_j||^2
                
                # Update koefisien dengan soft-thresholding
                if z > 0:  # Hindari pembagian oleh nol
                    w[j] = self.__soft_threshold(rho, self.alpha / 2) / z
                else:
                    w[j] = 0.0
            
            # Cek konvergensi setelah update semua fitur
            if np.max(np.abs(w - w_old)) < self.tol:
                print(f"Konvergen pada iterasi {iteration + 1}")
                break
        else:
            print(f"Peringatan: Tidak konvergen setelah {self.max_iter} iterasi")
        
        self.weight = w
        self.bias = b
        return self

    def predict(self, X):
        """Prediksi target untuk sampel baru
        
        Args:
            X (array-like): Matriks fitur dengan bentuk (n_samples, n_features)
        
        Returns:
            np.ndarray: Prediksi target dengan bentuk (n_samples,)
        
        Raises:
            ValueError: Jika model belum di-fit (weight=None)
            ValueError: Jika jumlah fitur tidak sesuai dengan training
        
        Examples:
            >>> model = LassoRegression().fit(X_train, y_train)
            >>> y_pred = model.predict(X_test)
        """
        if self.weight is None:
            raise ValueError("Model belum di-fit. Panggil .fit() terlebih dahulu.")
        
        X = np.asarray(X, dtype=float)
        if X.ndim == 1:
            X = X.reshape(1, -1)
        
        if X.shape[1] != self.weight.shape[0]:
            raise ValueError(
                f"Jumlah fitur tidak sesuai: training punya {self.weight.shape[0]} "
                f"fitur, prediksi punya {X.shape[1]} fitur"
            )
        
        return X @ self.weight + self.bias
```

## Elasticnet Regression

```python
import numpy as np

class ElasticNetRegression:
    """Elastic Net Regression dengan Coordinate Descent
    
    Menggabungkan L1 (Lasso) dan L2 (Ridge) regularization untuk:
    - Seleksi fitur otomatis (via L1)
    - Stabilitas pada fitur berkorelasi tinggi (via L2)
    
    Fungsi loss yang diminimalkan:
        (1/(2n)) * ||y - Xw - b||^2 
        + alpha * l1_ratio * ||w||_1 
        + (alpha * (1 - l1_ratio) / 2) * ||w||^2_2
    
    di mana:
        - ||w||_1 = sum(|w_j|) adalah L1 penalty (sparse)
        - ||w||^2_2 = sum(w_j^2) adalah L2 penalty (shrinkage)
        - alpha: kekuatan regularisasi keseluruhan
        - l1_ratio: mixing parameter (0 = Ridge murni, 1 = Lasso murni)
        - Intercept (b) TIDAK diregularisasi
    
    Attributes:
        alpha (float): Kekuatan regularisasi keseluruhan. Default: 1.0
        l1_ratio (float): Proporsi L1 penalty (0 ≤ l1_ratio ≤ 1). Default: 0.5
        fit_intercept (bool): Jika True, hitung intercept. Default: True
        max_iter (int): Maksimum iterasi Coordinate Descent. Default: 1000
        tol (float): Toleransi konvergensi. Default: 1e-4
        coef_ (np.ndarray): Koefisien fitur setelah training. Bentuk (n_features,)
        intercept_ (float): Intercept setelah training. 0.0 jika fit_intercept=False
    
    Notes:
        - **WAJIB standardisasi fitur** (mean=0, std=1) sebelum training
        - Pada multikolinearitas sempurna, Elastic Net membagi koefisien secara adil
        - Lebih stabil daripada Lasso pada fitur highly correlated
        - Tidak memiliki solusi closed-form → HARUS menggunakan optimisasi iteratif
    
    Examples:
        >>> import numpy as np
        >>> from sklearn.preprocessing import StandardScaler
        >>> 
        >>> # Dataset dengan multikolinearitas (X2 = 2*X1)
        >>> X = np.array([[1, 2], [2, 4], [3, 6], [4, 8]])
        >>> y = np.array([3, 5, 7, 9])
        >>> 
        >>> # Standardisasi WAJIB
        >>> scaler = StandardScaler()
        >>> X_scaled = scaler.fit_transform(X)
        >>> 
        >>> model = ElasticNetRegression(alpha=1.0, l1_ratio=0.5, max_iter=1000)
        >>> model.fit(X_scaled, y)
        Konvergen pada iterasi 8
        >>> 
        >>> print("Koefisien:", model.coef_)
        Koefisien: [1.0 1.0]  # Kedua fitur aktif dengan koefisien seimbang
        >>> print("Intercept:", model.intercept_)
        Intercept: 6.0
        >>> print("Prediksi:", model.predict(X_scaled))
        Prediksi: [3. 5. 7. 9.]
    """
    
    def __init__(self, alpha=1.0, l1_ratio=0.5, fit_intercept=True, 
                 max_iter=1000, tol=1e-4):
        """Inisialisasi Elastic Net Regression
        
        Args:
            alpha (float): Kekuatan regularisasi keseluruhan. Harus >= 0.
                - alpha = 0 → setara dengan OLS (tanpa regularisasi)
                - alpha besar → koefisien lebih kecil/sparse
            l1_ratio (float): Mixing parameter antara L1 dan L2 (0 ≤ l1_ratio ≤ 1).
                - l1_ratio = 1.0 → Lasso murni
                - l1_ratio = 0.0 → Ridge murni
                - l1_ratio = 0.5 → kombinasi 50-50 (default)
            fit_intercept (bool): Jika True, hitung intercept secara terpisah 
                (tidak diregularisasi). Default: True
            max_iter (int): Maksimum iterasi Coordinate Descent. Default: 1000
            tol (float): Toleransi untuk kriteria konvergensi. Iterasi berhenti 
                jika max(|w_new - w_old|) < tol. Default: 1e-4
        
        Raises:
            ValueError: Jika alpha < 0 atau l1_ratio tidak di [0, 1]
        
        Examples:
            >>> # Lasso murni
            >>> model = ElasticNetRegression(alpha=1.0, l1_ratio=1.0)
            >>> 
            >>> # Ridge murni
            >>> model = ElasticNetRegression(alpha=1.0, l1_ratio=0.0)
            >>> 
            >>> # Elastic Net standar
            >>> model = ElasticNetRegression(alpha=1.0, l1_ratio=0.5)
        """
        if alpha < 0:
            raise ValueError(f"alpha harus >= 0, ditemukan {alpha}")
        if not 0.0 <= l1_ratio <= 1.0:
            raise ValueError(f"l1_ratio harus di [0, 1], ditemukan {l1_ratio}")
        
        self.alpha = alpha
        self.l1_ratio = l1_ratio
        self.fit_intercept = fit_intercept
        self.max_iter = max_iter
        self.tol = tol
        self.coef_ = None
        self.intercept_ = 0.0

    def __soft_threshold(self, rho, threshold):
        """Operator soft-thresholding untuk komponen L1
        
        Fungsi proximal dari penalty L1:
            S(rho, threshold) = sign(rho) * max(|rho| - threshold, 0)
        
        Args:
            rho (float): Partial correlation antara fitur dan residual
            threshold (float): Threshold parameter (alpha * l1_ratio * n_samples)
        
        Returns:
            float: Nilai setelah soft-thresholding
        
        Examples:
            >>> model = ElasticNetRegression()
            >>> model._ElasticNetRegression__soft_threshold(3.0, 1.0)
            2.0
            >>> model._ElasticNetRegression__soft_threshold(-3.0, 1.0)
            -2.0
            >>> model._ElasticNetRegression__soft_threshold(0.5, 1.0)
            0.0
        """
        if rho > threshold:
            return rho - threshold
        elif rho < -threshold:
            return rho + threshold
        else:
            return 0.0

    def fit(self, X, y):
        """Fit model Elastic Net menggunakan Coordinate Descent
        
        Algoritma Coordinate Descent untuk Elastic Net:
            1. Inisialisasi koefisien = 0
            2. Untuk setiap iterasi:
                a. Update intercept (jika fit_intercept=True)
                b. Untuk setiap fitur j:
                    - Hitung partial residual: r = y - b - Xw + w_j * X_j
                    - Hitung partial correlation: rho = X_j^T * r
                    - Hitung norm kuadrat fitur: z = X_j^T * X_j
                    - Update koefisien:
                          w_j = S(rho, alpha * l1_ratio * n) / (z + alpha * (1-l1_ratio) * n)
                c. Cek konvergensi: max|w_new - w_old| < tol
        
        Args:
            X (array-like): Matriks fitur dengan bentuk (n_samples, n_features)
            y (array-like): Vektor target dengan bentuk (n_samples,)
        
        Returns:
            self: Instance model yang telah di-fit
        
        Raises:
            ValueError: Jika X atau y memiliki bentuk tidak valid
            TypeError: Jika input tidak dapat dikonversi ke float
        
        Notes:
            - Input 1D untuk X akan otomatis diubah menjadi (n_samples, 1)
            - Intercept tidak pernah diregularisasi
            - Standardisasi fitur WAJIB untuk hasil optimal
            - Konvergensi dicetak ke stdout jika tercapai sebelum max_iter
        
        Mathematical Derivation:
            Update rule untuk fitur j:
                w_j = S(rho, λ₁) / (z + λ₂)
            di mana:
                ρ = X_j^T (y - b - X_{-j} w_{-j})
                z = X_j^T X_j
                λ₁ = alpha * l1_ratio * n_samples   (L1 threshold)
                λ₂ = alpha * (1 - l1_ratio) * n_samples (L2 penalty)
        """
        # Validasi dan konversi input
        X = np.asarray(X, dtype=float)
        y = np.asarray(y, dtype=float).flatten()
        
        # Handle input 1D (vektor)
        if X.ndim == 1:
            X = X.reshape(-1, 1)
        
        if X.shape[0] != y.shape[0]:
            raise ValueError(
                f"Dimensi tidak sesuai: X punya {X.shape[0]} sampel, "
                f"y punya {y.shape[0]} sampel"
            )
        
        n_samples, n_features = X.shape
        w = np.zeros(n_features)
        
        # Inisialisasi intercept
        if self.fit_intercept:
            b = np.mean(y)
        else:
            b = 0.0
        
        # Precompute scaling factors untuk efisiensi
        l1_threshold = self.alpha * self.l1_ratio * n_samples
        l2_penalty = self.alpha * (1.0 - self.l1_ratio) * n_samples
        
        # Coordinate Descent
        for iteration in range(self.max_iter):
            w_old = w.copy()
            
            # Update intercept (tidak diregularisasi)
            if self.fit_intercept:
                b = np.mean(y - X @ w)
            
            # Update setiap koefisien fitur
            for j in range(n_features):
                # Hitung partial residual: residual tanpa kontribusi fitur j
                # r = y - b - (Xw - w_j * X_j) = y - b - Xw + w_j * X_j
                residual = y - b - (X @ w) + w[j] * X[:, j]
                
                # Hitung partial correlation dan norm kuadrat fitur
                rho = np.dot(X[:, j], residual)
                z = np.dot(X[:, j], X[:, j])  # = ||X_j||^2
                
                # Update koefisien dengan soft-thresholding + L2 penalty
                if z > 0:  # Hindari pembagian oleh nol
                    w[j] = self.__soft_threshold(rho, l1_threshold) / (z + l2_penalty)
                else:
                    w[j] = 0.0
            
            # Cek konvergensi setelah update semua fitur
            max_change = np.max(np.abs(w - w_old))
            if max_change < self.tol:
                print(f"Konvergen pada iterasi {iteration + 1} (max_change={max_change:.2e})")
                break
        else:
            print(f"Peringatan: Tidak konvergen setelah {self.max_iter} iterasi "
                  f"(max_change terakhir={max_change:.2e})")
        
        self.coef_ = w
        self.intercept_ = b
        return self

    def predict(self, X):
        """Prediksi target untuk sampel baru
        
        Args:
            X (array-like): Matriks fitur dengan bentuk (n_samples, n_features)
        
        Returns:
            np.ndarray: Prediksi target dengan bentuk (n_samples,)
        
        Raises:
            ValueError: Jika model belum di-fit (coef_=None)
            ValueError: Jika jumlah fitur tidak sesuai dengan training
        
        Examples:
            >>> model = ElasticNetRegression().fit(X_train, y_train)
            >>> y_pred = model.predict(X_test)
            >>> print(y_pred.shape)
            (n_test_samples,)
        """
        if self.coef_ is None:
            raise ValueError("Model belum di-fit. Panggil .fit() terlebih dahulu.")
        
        X = np.asarray(X, dtype=float)
        if X.ndim == 1:
            X = X.reshape(1, -1)
        
        if X.shape[1] != self.coef_.shape[0]:
            raise ValueError(
                f"Jumlah fitur tidak sesuai: training punya {self.coef_.shape[0]} "
                f"fitur, prediksi punya {X.shape[1]} fitur"
            )
        
        return X @ self.coef_ + self.intercept_
```

## KNN - algorithm

```python
import numpy as np
from collections import Counter


class KNearestNeighbor():
    """
    K-Nearest Neighbor (KNN) Classifier.

    Algoritma supervised learning sederhana yang bekerja dengan cara:
    1. Menghitung jarak data baru ke seluruh data training.
    2. Mengambil k tetangga terdekat.
    3. Menentukan label berdasarkan voting mayoritas.

    Parameters
    ----------
    k : int, default=5
        Jumlah tetangga terdekat yang digunakan untuk voting.
    """

    def __init__(self, k=5):
        self.k = k
        self.x_train = None
        self.y_train = None

    def fit(self, X, y):
        """
        Menyimpan data training.

        Parameters
        ----------
        X : array-like, shape (n_samples, n_features)
            Data fitur training.
        y : array-like, shape (n_samples,)
            Label training.
        """
        self.x_train = np.array(X)
        self.y_train = np.array(y)

    def predict(self, X):
        """
        Memprediksi label untuk data baru.

        Parameters
        ----------
        X : array-like, shape (n_samples, n_features)
            Data fitur yang ingin diprediksi.

        Returns
        -------
        numpy.ndarray
            Array berisi hasil prediksi label.
        """
        X = np.array(X)
        predictions = [self._predict_single_sample(x) for x in X]
        return np.array(predictions)

    def _predict_single_sample(self, x):
        """
        Memprediksi label untuk satu sampel data.

        Langkah-langkah:
        1. Hitung jarak Euclidean ke seluruh data training.
        2. Ambil k indeks dengan jarak terdekat.
        3. Lakukan voting mayoritas.

        Parameters
        ----------
        x : array-like
            Satu data fitur yang ingin diprediksi.

        Returns
        -------
        int / str
            Label hasil voting mayoritas.
        """
        distances = np.sqrt(np.sum((self.x_train - x) ** 2, axis=1))

        k_indices = np.argsort(distances)[:self.k]

        k_nearest_labels = [self.y_train[i] for i in k_indices]

        most_common = Counter(k_nearest_labels).most_common(1)

        return most_common[0][0]
```

## Weighted-KNN - algorithm

```python
import numpy as np

class WeightedKNearestNeighbor:
    """
    Weighted K-Nearest Neighbor Classifier.

    Parameters
    ----------
    k : int
        Jumlah tetangga terdekat.
    weight_scheme : str
        Skema pembobotan:
        - 'inverse_distance'
        - 'gaussian'
        - 'exponential'
    """

    def __init__(self, k=5, weight_scheme='inverse_distance'):
        self.k = k
        self.weight_scheme = weight_scheme
        self.x_train = None
        self.y_train = None

    def fit(self, x, y):
        self.x_train = np.array(x)
        self.y_train = np.array(y)
        return self

    def _calculate_weights(self, distances):
        if self.weight_scheme == 'inverse_distance':
            weights = 1.0 / np.where(distances == 0, 1e-10, distances)

        elif self.weight_scheme == 'gaussian':
            sigma = np.median(distances) + 1e-10
            weights = np.exp(-(distances ** 2) / (2 * sigma ** 2))

        elif self.weight_scheme == 'exponential':
            weights = np.exp(-distances)

        else:
            raise ValueError("Invalid weight scheme")

        return weights

    def predict(self, x):
        x = np.array(x)
        predictions = []

        for query in x:
            distances = np.sum((self.x_train - query) ** 2, axis=1)

            k_indices = np.argsort(distances)[:self.k]
            k_distances = distances[k_indices]
            k_labels = self.y_train[k_indices]

            weights = self._calculate_weights(k_distances)

            unique_labels = np.unique(k_labels)
            weighted_votes = {}

            for label in unique_labels:
                mask = (k_labels == label)
                weighted_votes[label] = np.sum(weights[mask])

            pred = max(weighted_votes, key=weighted_votes.get)
            predictions.append(pred)

        return np.array(predictions)
```

## NN - algorithm

```python
class NeuralNetworkRegression():
    """
    Implementasi sederhana Neural Network untuk regresi
    menggunakan satu hidden layer dan aktivasi ReLU.

    Arsitektur:
    Input -> Linear -> ReLU -> Linear -> Output

    Loss function:
    Mean Squared Error (MSE)

    Optimizer:
    Gradient Descent
    """

    def __init__(self, input_size, hidden_size, output_size=1, lr=0.01):
        """
        Inisialisasi parameter jaringan.

        Parameters
        ----------
        input_size : int
            Jumlah fitur input.
        hidden_size : int
            Jumlah neuron pada hidden layer.
        output_size : int, default=1
            Jumlah output (untuk regresi biasanya 1).
        lr : float, default=0.01
            Learning rate untuk update bobot.
        """
        self.lr = lr
        self.W1 = np.random.randn(input_size, hidden_size) * np.sqrt(1. / input_size)
        self.b1 = np.zeros((1, hidden_size))
        self.W2 = np.random.randn(hidden_size, output_size) * np.sqrt(1. / hidden_size)
        self.b2 = np.zeros((1, output_size))

    def relu(self, x):
        """
        Fungsi aktivasi ReLU.

        Parameters
        ----------
        x : ndarray
            Input array.

        Returns
        -------
        ndarray
            Hasil aktivasi ReLU.
        """
        return np.maximum(0, x)

    def relu_derivative(self, x):
        """
        Turunan fungsi ReLU.

        Parameters
        ----------
        x : ndarray
            Input sebelum aktivasi.

        Returns
        -------
        ndarray
            Gradien ReLU terhadap input.
        """
        return (x > 0).astype(float)

    def forward(self, X):
        """
        Melakukan forward propagation.

        Parameters
        ----------
        X : ndarray, shape (n_samples, input_size)

        Returns
        -------
        ndarray
            Prediksi output jaringan.
        """
        self.Z1 = X @ self.W1 + self.b1
        self.A1 = self.relu(self.Z1)
        self.Z2 = self.A1 @ self.W2 + self.b2
        return self.Z2

    def compute_loss(self, y_true, y_pred):
        """
        Menghitung Mean Squared Error (MSE).

        Parameters
        ----------
        y_true : ndarray
            Target asli.
        y_pred : ndarray
            Prediksi model.

        Returns
        -------
        float
            Nilai loss.
        """
        return np.mean((y_true - y_pred) ** 2)

    def backward(self, X, y_true):
        """
        Melakukan backpropagation dan update bobot.

        Parameters
        ----------
        X : ndarray
            Data input.
        y_true : ndarray
            Target asli.
        """
        n = X.shape[0]
        y_pred = self.Z2

        dZ2 = (2 / n) * (y_pred - y_true)
        dW2 = self.A1.T @ dZ2
        db2 = np.sum(dZ2, axis=0, keepdims=True)

        dA1 = dZ2 @ self.W2.T
        dZ1 = dA1 * self.relu_derivative(self.Z1)
        dW1 = X.T @ dZ1
        db1 = np.sum(dZ1, axis=0, keepdims=True)

        self.W2 -= self.lr * dW2
        self.b2 -= self.lr * db2
        self.W1 -= self.lr * dW1
        self.b1 -= self.lr * db1

    def fit(self, X, y, epochs=1000):
        """
        Melatih model menggunakan Gradient Descent.

        Parameters
        ----------
        X : ndarray
            Data training.
        y : ndarray
            Target training.
        epochs : int, default=1000
            Jumlah iterasi pelatihan.

        Returns
        -------
        self
        """
        X = np.array(X)
        y = np.array(y)

        if y.ndim == 1:
            y = y.reshape(-1, 1)

        for _ in range(epochs):
            self.forward(X)
            self.backward(X, y)

        return self

    def predict(self, X):
        """
        Melakukan prediksi pada data baru.

        Parameters
        ----------
        X : ndarray
            Data yang ingin diprediksi.

        Returns
        -------
        ndarray
            Hasil prediksi.
        """
        return self.forward(np.array(X))
```
