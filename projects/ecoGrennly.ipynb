## ðŸ’» **Implementasi Step-by-Step (Python + Scikit-learn)**

### ðŸ”‘ **Step 1: Bangun User-Item Matrix**
```python
import pandas as pd
import numpy as np
from scipy.sparse import csr_matrix

# Load data
orders = pd.read_csv('olist_orders_dataset.csv')
items = pd.read_csv('olist_order_items_dataset.csv')
reviews = pd.read_csv('olist_order_reviews_dataset.csv')

# Gabungkan menjadi user-item interactions
df = orders.merge(items, on='order_id').merge(reviews[['order_id', 'review_score']], on='order_id', how='left')

# Isi missing ratings dengan rata-rata (3.0)
df['review_score'] = df['review_score'].fillna(3.0)

# Pivot ke user-item matrix
user_item_matrix = df.pivot_table(
    index='customer_id',
    columns='product_id',
    values='review_score',
    aggfunc='mean'
).fillna(0)  # 0 = tidak ada interaksi

print(f"Matriks: {user_item_matrix.shape[0]} users Ã— {user_item_matrix.shape[1]} products")
```

### ðŸ”‘ **Step 2: Collaborative Filtering dengan Matrix Factorization (SVD)**
```python
from sklearn.decomposition import TruncatedSVD
from sklearn.preprocessing import MinMaxScaler

# Konversi ke sparse matrix
sparse_matrix = csr_matrix(user_item_matrix.values)

# Matrix Factorization (SVD)
svd = TruncatedSVD(n_components=50, random_state=42)
user_factors = svd.fit_transform(sparse_matrix)
item_factors = svd.components_.T

# Prediksi ratings
predicted_ratings = np.dot(user_factors, item_factors)

# Normalisasi ke skala 1-5
scaler = MinMaxScaler(feature_range=(1, 5))
predicted_ratings = scaler.fit_transform(predicted_ratings)

print("Shape faktor user:", user_factors.shape)
print("Shape faktor item:", item_factors.shape)
```

### ðŸ”‘ **Step 3: Content-Based Filtering (Product Similarity)**
```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

products = pd.read_csv('olist_products_dataset.csv')
category_trans = pd.read_csv('product_category_name_translation.csv')

# Gabungkan kategori Portuguese â†’ English
products = products.merge(category_trans, on='product_category_name', how='left')

# Buat "deskripsi" dari atribut produk
products['product_features'] = (
    products['product_category_name_english'].fillna('') + ' ' +
    products['product_name_lenght'].astype(str) + ' ' +
    products['product_weight_g'].astype(str)
)

# TF-IDF vectorization
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(products['product_features'].fillna(''))

# Hitung similarity antar produk
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Fungsi rekomendasi berdasarkan produk serupa
def get_content_recommendations(product_id, n=5):
    idx = products[products['product_id'] == product_id].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:n+1]  # Skip diri sendiri
    product_indices = [i[0] for i in sim_scores]
    return products.iloc[product_indices][['product_id', 'product_category_name_english']]
```

### ðŸ”‘ **Step 4: Hybrid Recommendation (Weighted Combination)**
```python
def hybrid_recommend(user_id, product_id, alpha=0.6):
    """
    alpha = 0.6 â†’ 60% collaborative, 40% content-based
    """
    # Collaborative filtering score
    user_idx = user_item_matrix.index.get_loc(user_id)
    collab_scores = predicted_ratings[user_idx]
    
    # Content-based score (kemiripan dengan produk yang dilihat)
    product_idx = products[products['product_id'] == product_id].index[0]
    content_scores = cosine_sim[product_idx]
    
    # Hybrid score
    hybrid_scores = (alpha * collab_scores) + ((1 - alpha) * content_scores)
    
    # Ambil top-N rekomendasi
    top_n_idx = np.argsort(hybrid_scores)[::-1][:10]
    recommended_products = user_item_matrix.columns[top_n_idx]
    
    return recommended_products.tolist()
```
