### Import tools

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler


def train_test_split(X, Y, test_size=0.2, random_state=None):
    if random_state is not None:
        np.random.seed(random_state)
    X = np.array(X)
    Y = np.array(Y)
    
    n = len(X)
    test_n = int(n * test_size)
    indices = np.random.permutation(n) 
    test_idx = indices[:test_n]
    train_idx = indices[test_n:]
    
    return X[train_idx], X[test_idx], Y[train_idx], Y[test_idx]

def get_mean(arr):
    return sum(arr)/len(arr)

def get_sd(arr):
    mean=get_mean(arr)
    variances=sum((x - mean) ** 2 for x in arr)/len(arr)

    return variances ** 0.5

def z_score(arr, mean, sd):
  if sd==0:
    return [0 for _ in arr]
  return [(x - mean)/sd for x in arr]


def mse(y_true, y_pred):
    return sum((yt - y_p) ** 2 for yt, y_p in zip(y_true, y_pred)) / len(y_true)

def rmse(y_true, y_pred):
    return mse(y_true, y_pred) ** 0.5

def mae(y_true, y_pred):
    return sum(abs(yt - yp) for yt, yp in zip(y_true, y_pred)) / len(y_true)

def r2_score(y_true, y_pred):
    ss_residual=sum((yt - yp) ** 2 for yt, yp in zip(y_true, y_pred))
    ss_total=sum((yt - get_mean(y_true)) ** 2 for yt in y_true)
    if ss_total == 0:
        return 1 if ss_residual == 0 else 0
    return 1 - (ss_residual / ss_total)

def adjusted_r2_score(y_true, y_pred, n_features):
    r2 = r2_score(y_true, y_pred)
    n = len(y_true)
    if n <= n_features:
        return 0
    return 1 - (1 - r2) * (n - 1) / (n - n_features - 1)

def one_hot_encode(col, ccategories):
    encoding = {}
    for category in ccategories:
        encoding[category] = [1 if val == category else 0 for val in col]
    return encoding

def label_encode(col):
    return [1 if x == 'yes' else 0 for x in col]
```

```python
class SimpleLinearRegression:
    def __init__(self):
        self.weight = None
        self.intercept = None

    def fit(self, X, Y):

        n=len(X)
        x_mean=get_mean(X)
        y_mean=get_mean(Y)

        self.weight = sum(
            (X[i] - x_mean) * (Y[i] - y_mean) for i in range(n)
        ) / sum(
            (X[i] - x_mean) ** 2 for i in range(n)
        )
        self.intercept = y_mean - self.weight * x_mean

    def predict(self, X):
        if self.weight is None or self.intercept is None:
            raise ValueError("Model belum dilatih! Panggil .fit() dulu.")
        return [self.intercept + self.weight * x for x in X]

    def evaluate(self, Y_true, Y_pred):
        return {
            'MSE': mse(Y_true, Y_pred),
            'RMSE': rmse(Y_true, Y_pred),
            'MAE': mae(Y_true, Y_pred),
            'R2': r2_score(Y_true, Y_pred)
        }

    def pipeline(self, X, Y, test_size=0.2, random_state=42, standardize=False):
        X_train, X_test, Y_train, Y_test = train_test_split(X, Y,test_size=test_size, random_state=random_state)

        if standardize:
            #  HANYA gunakan statistik dari X_train!
            x_mean = get_mean(X_train)
            x_sd = get_sd(X_train)
            X_train = z_score(X_train, x_mean, x_sd)
            X_test = z_score(X_test, x_mean, x_sd) 

        self.fit(X_train, Y_train)

        train_metrics= self.evaluate(Y_train, self.predict(X_train))
        test_metrics= self.evaluate(Y_test, self.predict(X_test))

        return{
            'train_metrics': train_metrics,
            'test_metrics': test_metrics,
            'intercept': self.intercept,
            'weight': self.weight
        }
```

```python
class MultipleLinearRegression:
    def __init__(self, fit_intercept=True):
        self.weight = None
        self.intercept = None
        self.fit_intercept = fit_intercept
    
    def fit(self, X, Y):
        X= np.array(X, dtype=float)
        Y= np.array(Y, dtype=float)

        if self.fit_intercept:
            X = np.column_stack((np.ones(X.shape[0]), X))
        
        XtX=np.dot(X.T, X)
        XtX_invers = np.linalg.inv(XtX)
        XtY=np.dot(X.T, Y)
        beta=np.dot(XtX_invers, XtY)

        if self.fit_intercept:
            self.intercept = beta[0]
            self.weight = beta[1:]
        else:
            self.intercept = 0
            self.weight = beta

        pass

    def predict(self, X):
        X = np.array(X, dtype=float)
        if self.weight is None:
            raise ValueError("Model belum dilatih! Panggil .fit() dulu.")
        if self.fit_intercept:
            return self.intercept + np.dot(X, self.weight)
        else:
            return np.dot(X, self.weight)   


    def evaluate(self, Y_true, Y_pred, n_features):
        y_true= np.array(Y_true, dtype=float)
        y_pred= np.array(Y_pred, dtype=float)

        sns.histplot(y_true - y_pred, kde=True)
        plt.title('Distribusi Residuals')
        plt.xlabel('Residuals')
        plt.ylabel('Frekuensi')
        plt.show()
        
        return {
            'MSE': mse(y_true, y_pred),
            'RMSE': rmse(y_true, y_pred),
            'MAE': mae(y_true, y_pred),
            'R2': r2_score(y_true, y_pred),
            'Adjusted R2': adjusted_r2_score(y_true, y_pred, n_features)
        }
    
    def pipeline(self, X, Y, test_size=0.2, random_state=42, standardize=False):
        # 1. Konversi input ke numpy array
        X = np.array(X, dtype=float)
        Y = np.array(Y, dtype=float)
        
        # 2. Split data
        X_train, X_test, Y_train, Y_test = train_test_split(
            X, Y, test_size=test_size, random_state=random_state
        )
        
        # Pastikan hasil split tetap numpy array
        X_train = np.array(X_train, dtype=float)
        X_test = np.array(X_test, dtype=float)
        Y_train = np.array(Y_train, dtype=float)
        Y_test = np.array(Y_test, dtype=float)
        
        # 3. Standarisasi (jika diminta)
        if standardize:
            X_train_scaled = np.copy(X_train)
            X_test_scaled = np.copy(X_test)
            for i in range(X_train.shape[1]):
                col_train = X_train[:, i]
                mean = get_mean(col_train)
                sd = get_sd(col_train)
                X_train_scaled[:, i] = z_score(col_train, mean, sd)
                X_test_scaled[:, i] = z_score(X_test[:, i], mean, sd)  # pakai statistik TRAIN!
            X_train, X_test = X_train_scaled, X_test_scaled
        
        # 4. Latih & prediksi
        self.fit(X_train, Y_train)
        Y_train_pred = self.predict(X_train)
        Y_test_pred = self.predict(X_test)
        
        # 5. Evaluasi
        n_feat = X_train.shape[1]  #  Sekarang X_train pasti numpy array!
        train_metrics = self.evaluate(Y_train, Y_train_pred, n_feat)
        test_metrics = self.evaluate(Y_test, Y_test_pred, n_feat)
        
        return {
            'train_metrics': train_metrics,
            'test_metrics': test_metrics,
            'intercept': self.intercept,
            'weights': self.weight,
            'n_features': n_feat
        }
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import numpy as np
import pandas as pd

def calculate_vif(X, feature_names):
    """
    Hitung VIF untuk setiap fitur.
    X: array-like, shape (n_samples, n_features)  tanpa intercept
    feature_names: list nama fitur (panjang = n_features)
    """
    X = np.array(X, dtype=float)
    
    #  Tambahkan kolom intercept (konstanta 1)
    X_with_const = np.column_stack([np.ones(X.shape[0]), X])
    
    #  Hitung VIF untuk setiap fitur (lewati kolom intercept di index 0)
    vif_values = []
    for i in range(X.shape[1]):
        vif = variance_inflation_factor(X_with_const, i + 1)  # +1 karena intercept di index 0
        vif_values.append(vif)
    
    #  Buat DataFrame dengan nama fitur yang benar
    vif_df = pd.DataFrame({
        'Feature': feature_names,
        'VIF': vif_values
    })
    
    return vif_df
```

```python

X = house_df_copy.drop('price', axis=1).values.tolist()
Y = [np.log(price) for price in house_df_copy['price'].tolist()]


vif_results = calculate_vif(X, house_df_copy.drop('price', axis=1).columns.tolist())
print(" Variance Inflation Factor (VIF) untuk Fitur:")
print(vif_results.sort_values(by='VIF', ascending=False))
```
