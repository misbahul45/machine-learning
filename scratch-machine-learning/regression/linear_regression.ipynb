```python
def one_hot_function(
   df_input,
   df_output,   
):
    df_encoded=pd.get_dummies(df_input, dtype=int, drop_first=True)
    return pd.concat([df_output, df_encoded], axis=1)
```

```python
one_hot_encoded_cols=['Job Roles', 'Employment Status', 'Location']

for col in one_hot_encoded_cols:
    df=one_hot_function(df[[col]], df)
    df.drop(columns=[col], inplace=True)
```
