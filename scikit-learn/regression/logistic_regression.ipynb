```python
df['diagnosis']=[1 if d =='M' else 0 for d in df['diagnosis']]
```

```python
df['diagnosis'].value_counts()
```

```python
Y=df['diagnosis']
X=df.drop(columns='diagnosis')
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

X_scaled=scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test=train_test_split(X_scaled, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

# create the logistic regression model
model=LogisticRegression()

model.fit(X_train, Y_train)

Y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report

print(f'accuracy {accuracy_score(Y_test, Y_pred)}')
print(classification_report(Y_test, Y_pred))
```

```python
import numpy as np
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression



# Buat pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('classifier', LogisticRegression())
])

# Gunakan StratifiedKFold untuk menjaga proporsi kelas
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Hitung ROC AUC via cross-validation
roc_auc_scores = cross_val_score(pipeline, X, Y, cv=cv, scoring='roc_auc')

print(f"ROC AUC (CV): {roc_auc_scores.mean():.3f} ± {roc_auc_scores.std():.3f}")
```
