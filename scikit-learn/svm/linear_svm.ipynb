```python
from sklearn.preprocessing import OneHotEncoder

categorical_cols = ['cb_person_default_on_file','person_home_ownership']  

encoder = OneHotEncoder(drop='first', sparse_output=False) 

encoded_array = encoder.fit_transform(df_regression[categorical_cols])

encoded_cols = encoder.get_feature_names_out(categorical_cols)

df_encoded = pd.DataFrame(encoded_array, index=df_regression.index, columns=encoded_cols)

numerical_cols = df_regression.drop(columns=categorical_cols).columns
df_regression = pd.concat([df_regression[numerical_cols], df_encoded], axis=1)
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

df_imputed = df_regression.copy()

# Hanya ambil kolom numerik (termasuk kolom yang missing)
numeric_cols = df_imputed.select_dtypes(include=[np.number]).columns.tolist()


pipeline=Pipeline([
    ('scaler', StandardScaler()),
    ('imputer', IterativeImputer(
        estimator=SGDRegressor(max_iter=1000, tol=1e-3, random_state=42),
        random_state=42,
        max_iter=10,
        verbose=0
    ))
])

df_imputed[numeric_cols] = pipeline.fit_transform(df_imputed[numeric_cols])

df_regression=df_imputed.copy()
```
