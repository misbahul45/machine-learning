```python
df=pd.concat([df_true, df_fake], ignore_index=True)
df=df.sample(frac=1, random_state=42).reset_index(drop=True)

df.head()
```

```python
df['content']=df['title']+' '+df['subject']+' '+df['text']
```

# Nlp text prepocessing

```python
import string
import re

def preprocessing(text):
    if not isinstance(text, str):
        return ""
    
    text=text.lower()
    
    text="".join([char for char in text if char not in string.punctuation])

    text = re.sub(r'[^\w\s]', '', text)  # Keeps alphanumeric + whitespace
    
    # Optional: Remove extra whitespace
    text = " ".join(text.split())

    return text
```

```python
df['content']=df['content'].apply(preprocessing)
```

```python
import spacy

# Muat model bahasa Inggris kecil (sudah termasuk tokenizer + POS tagger + lemmatizer)
nlp = spacy.load("en_core_web_sm")

def lemmatize_text(text):
    if not isinstance(text, str):
        return ""
    # Proses teks dengan spaCy
    doc = nlp(text)
    # Ambil lemma dari setiap token yang bukan stopword/spasi (opsional: sesuaikan kebutuhan)
    lemmatized = [token.lemma_ for token in doc if not token.is_space]
    return " ".join(lemmatized)
```

```python
df['content']=df['content'].apply(lemmatize_text)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

X=df['content']
Y=df['label']

vectorizer=TfidfVectorizer()

X_train, X_test, Y_train, Y_test=train_test_split(X, Y, test_size=0.2, random_state=42, stratify=Y)

X_train=vectorizer.fit_transform(X_train)
X_test=vectorizer.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB

model=MultinomialNB()

model.fit(X_train, Y_train)

Y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(Y_test, Y_pred))
```
