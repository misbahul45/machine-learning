```python

x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)

fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])

ax.scatter(x, y, color="yellow", label='sin(x)')
ax.scatter(x, z, color='black', label='cos(x)')

# tampilkan nilai (x, y) untuk sin
for i in range(len(x)):
    ax.annotate(
        f"({x[i]:.2f}, {y[i]:.2f})",
        (x[i], y[i]),
        fontsize=6,
        alpha=0.7
    )

# tampilkan nilai (x, z) untuk cos
for i in range(len(x)):
    ax.annotate(
        f"({x[i]:.2f}, {z[i]:.2f})",
        (x[i], z[i]),
        fontsize=6,
        alpha=0.7
    )

ax.legend()
plt.show()
```

```python
from itertools import combinations

numeric_cols = diabetes_df.select_dtypes(include='number').columns


for col_x, col_y in combinations(numeric_cols, 2):

    x = diabetes_df[col_x]
    y = diabetes_df[col_y]

    # regresi linear
    m, c = np.polyfit(x, y, 1)
    y_pred = m * x + c

    # plot
    fig, ax = plt.subplots()

    ax.scatter(x, y, alpha=0.6, label='Data points')
    ax.plot(x, y_pred, label='Regression line')

    ax.set_xlabel(col_x)
    ax.set_ylabel(col_y)
    ax.set_title(f'{col_x} vs {col_y}')
    ax.legend()

    plt.show()
```

```python
tips=pd.read_csv('./data/tip.csv')

print(tips.head())
```

```python
# Relational plot
g = sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    col="time",
    hue="smoker",
    size="size",
    kind="scatter",
    height=4,      # tinggi tiap subplot
    aspect=1.2     # lebar relatif
)

# Judul utama
g.fig.suptitle("Hubungan Total Bill vs Tip", fontsize=14, y=1.05)

# Rapikan layout
g.fig.tight_layout()

plt.show()
```

```python
numeric_titanic_df = titanic_df.select_dtypes(include='number')
corr_matrix = numeric_titanic_df.corr()
```

```python
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

plt.figure(figsize=(10, 8))
sns.heatmap(
    corr_matrix,
    mask=mask,
    annot=True,
    fmt=".2f",
    cmap='coolwarm'
)
plt.title("Correlation Heatmap (Lower Triangle)")
plt.show()
```
